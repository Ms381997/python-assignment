2. Write a program that asks the user for the side of a square and prints its perimeter.

```python
# Program to calculate the perimeter of a square

# 1. Ask the user for the side length
side_input = input("Enter the side of the square: ")

# 2. Convert the input to a float (to allow for decimals)
side = float(side_input)

# 3. Calculate the perimeter (4 * side)
perimeter = 4 * side

# 4. Print the result
print("The perimeter of the square is:", perimeter)
```

3. Take the base and height of a triangle as input and print its area.

```python
# 1. Take base and height as input from the user
base = float(input("Enter the base of the triangle: "))
height = float(input("Enter the height of the triangle: "))

# 2. Calculate the area using the formula
area = 0.5 * base * height

# 3. Print the result
print(f"The area of the triangle is: {area}")
```

4. Write a program that asks the user for the radius of a circle and prints its circumference. (Use 3.14 for π).

```python
# Ask the user for the radius
radius_input = input("Enter the radius of the circle: ")

# Convert the input string to a float for calculation
radius = float(radius_input)

# Use 3.14 for pi
pi = 3.14

# Calculate circumference: 2 * pi * r
circumference = 2 * pi * radius

# Print the result
print("The circumference of the circle is:", circumference)
```

5. Take Principal (P), Rate (R), and Time (T) as input from the user and print the Simple Interest.

```python
# Taking inputs
P = float(input("Enter the Principal amount: "))
R = float(input("Enter the Rate of interest: "))
T = float(input("Enter the Time (in years): "))

# Calculating Simple Interest
SI = (P * R * T) / 100

# Printing the result
print(f"The Simple Interest is: {SI}")
```
